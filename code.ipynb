```python
from sklearn.model_selection import KFold,cross_val_score
from sklearn.linear_model import LinearRegression
x=df.drop('SalePrice',axis=1)
y=df['SalePrice']

model=LinearRegression()

kf=KFold(n_splits=5,shuffle=True)

scores=cross_val_score(model,x,y,cv=kf,scoring='r2')

print("scores for each fold:", scores)
print("Mean score:", np.mean(scores))
print("Min score:", np.min(scores))
print("Max score:", np.max(scores))
```

4 . Other models 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Load and preprocess the data
df = pd.read_csv('train.csv')
columns_isnull = [i for i in df.columns if df[i].isnull().sum() > 600]
df.drop(columns=columns_isnull, inplace=True)

cols_same = []
for col in df.columns:
    most = df[col].mode()[0]
    count_col = (df[col] == most).sum()
    if count_col / len(df) > 0.8:
        cols_same.append(col)
df.drop(columns=cols_same, inplace=True)

categorical_col = [column for column in df.columns if df[column].dtype == object]
label = LabelEncoder()
for col in categorical_col:
    df[col] = label.fit_transform(df[col])

df.fillna(method='ffill', inplace=True)

# Split the data into training and testing sets
x = df.drop('SalePrice', axis=1)
y = df['SalePrice']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1)

# Define the models and their parameter grids
models = { 
    'Ridge': {
        'model': Ridge(),
        'params': {
            'alpha': [0.1, 1, 10, 100],
            'fit_intercept': [True, False],
            
        }
    },
    'Lasso': {
        'model': Lasso(),
        'params': {
            'alpha': [0.1, 1, 10, 100],
            'fit_intercept': [True, False],
            
        }
    },
    'RandomForestRegressor': {
        'model': RandomForestRegressor(),
        'params': {
            'n_estimators': [10, 50, 100],
            'max_features': ['auto', 'sqrt', 'log2'],
            'max_depth': [None, 10, 20, 30]
        }
    }
}

# Use GridSearchCV to find the best model and parameters
scores = []
for model_name, mp in models.items():
    clf = GridSearchCV(mp['model'], mp['params'], cv=5, scoring='r2', return_train_score=False)
    clf.fit(x_train, y_train)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })

best_model_info = max(scores, key=lambda x: x['best_score'])
best_model = models[best_model_info['model']]['model']
best_model.set_params(**best_model_info['best_params'])
best_model.fit(x_train, y_train)
y_pred = best_model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2 = best_model.score(x_test, y_test)

print(f"Best Model: {best_model_info['model']}")
print(f"Best Parameters: {best_model_info['best_params']}")
print(f"The Mean Square Error(MSE) is: {mse}")
print(f"R square obtained is: {r2}")

# Optional: Plotting the actual vs predicted values
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title(f"{best_model_info['model']} - Actual vs Predicted Prices")
plt.show()

```
